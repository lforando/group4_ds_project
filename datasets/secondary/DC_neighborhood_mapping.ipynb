```python
col_list = ['PROPTYPE', 
            'NBHDNAME',
           'ASSESSMENT']

df = pd.read_csv('property_info.csv',
                usecols = col_list)
```

```python
neighborhood_names = df.NBHDNAME.unique()
neighborhood_count = []
for n in neighborhood_names:
    temp = df[df.NBHDNAME==n]
    count = len(temp)
    neighborhood_count.append(count)
    med_val = np.median(temp.ASSESSMENT)
    print(f'{n}: \n\t${med_val}')
    
nbh_df = pd.DataFrame(data={'nbhd': neighborhood_names,
                      'nb_count': neighborhood_count})

nbh_df = nbh_df.sort_values(by=['nb_count'], ascending=False)
```

```python
figure = plt.figure(figsize=(20, 20)) # first element is width, second is height.

y_pos = np.arange(len(neighborhood_names))
axes = figure.add_subplot(1, 1, 1)

axes.barh(y_pos, nbh_df.nb_count, color="dimgray")
axes.set_title( "Real Estate Counts for Washington DC, By Neighborhood")
axes.set_yticks(y_pos)
axes.set_yticklabels(neighborhood_names)
axes.invert_yaxis()
axes.set_xlabel('Count')



plt.show()
plt.close()
```
